## Interpretation of Loss Curves

The training loss decreased steadily from **0.5592** to **0.4336**, showing that the classifier learned useful patterns from the frozen ESM-2 embeddings.

The validation loss also improved overall, reaching its best value at epoch 8. After epoch 8, validation loss increased slightly while training loss continued to decrease. This suggests the model had largely converged by epoch 8, and additional epochs provided limited validation improvement.

Overall, the loss curves show stable learning with no severe overfitting.


## Interpretation of Accuracy Curves

Training accuracy increased from **69.45%** to **78.72%**.

Validation accuracy increased from **74.85%** to **76.97%** and then plateaued near the end of training.

The final train-validation accuracy gap was small:

\[
78.72\% - 76.97\% = 1.75\%
\]

This small gap suggests good generalization and limited overfitting.


## Research Interpretation

This run is a successful first ESM-2 baseline.

The model uses frozen ESM-2 embeddings with a relatively simple 1D CNN classifier, yet still reaches approximately **77% validation accuracy** on the solubility task. This suggests that ESM-2 embeddings contain useful biological information for downstream protein classification.

Because the model converges cleanly and does not show severe overfitting, this run can serve as an initial baseline for later comparisons against:

- autoencoder-based representations,
- random mutation augmentation,
- conservative/BLOSUM mutation augmentation,
- latent-space decoder augmentation.


## Limitations of Current Result
